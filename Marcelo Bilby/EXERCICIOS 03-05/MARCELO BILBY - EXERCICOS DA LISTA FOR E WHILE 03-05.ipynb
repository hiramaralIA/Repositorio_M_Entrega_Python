<h1>2 - Enquanto o usuário não encerrar o programa, significa que ele está registrando novos produtos</h1>
<h2>Registro de venda</h2>

```python
registro_venda = []
sair = True
```

```python
produto = input("Informe o produto que deve ser adicionado na lista de venda: ")
registro_venda.append(produto)
while sair != False:
  status = input("Você ainda deseja adicionar um produto: S/N ")
  if len(status) == 1 and status.isalpha():
    if status.upper() == "S" or status.upper() == "N":
        if status.upper() == "S":
          produto = input("Informe o produto que deve ser adicionado na lista de venda: ")
          registro_venda.append(produto)
        else:
          sair = False
    else:
      print("Informe um valor válido: ")
  else: 
    print("Informe um valor válido: ")
print(registro_venda)
```

<h1>3 - Registrar produtos e quantidades e adicionar a uma lista, continuar adicionando até o valor do produto igual a vazio</h1>

```python
registro_compra = []
```

```python
produto = input("Informe o produto a ser registrado: ")
qtde = int(input("Informe a quantidade: "))
while produto != "":
  registro_compra.append((produto, qtde))
  produto = input("Informe o produto a ser registrado: ")
  if produto != "":
    qtde = int(input("Informe a quantidade: "))
print(registro_compra)
```

<h1>7 - Crie um algoritmo que mostre os número pares de 1 a 50</h1> 

```python
print("Os valores em verde são os números pares.")
for i in range(1, 51):
  resto = i % 2
  if resto == 0:
    print(f"\033[32m {i} ", end=" ")
  else: 
    print(f"\033[31m {i}", end= " ")
```

<h1>8 - Crie um Algoritmo que calcule a soma de todos os número ímpares de 0 até 500</h1>

```python
impares = []
```

```python
for i in range(1, 500):
  resto = i % 2
  if resto != 0:
    impares.append(i)

soma = sum(impares)

print(f"A soma dos valores impares é: {soma}")
```

<h1>9 - Desenvolva um programa que leia seis números inteiros e mostre a soma apenas dos números pares</h1>

```python
soma = 0
for i in range(1, 7):
  numero = int(input(f"Informe o {i}° número: "))
  if numero % 2 == 0:
    soma += numero;
print(soma)
```

<h1>10 - Faça um algoritmo que leia um número inteiro e diga se ele é ou não um número primo</h1>

```python
num = int(input("Informe um número: "))

if num > 1:
    for i in range(2, num+1):
        if num == 2:
          print(num, "é um número primo")

        else:
            if num % i == 0:
              print(num, "Não é número primo") 
              break
            else:
              print(num, "é um número primo")
              break
else:
    print(num, "não é um número primo")
```
