```python
quarto = []
```

```python
quantidade = int(input("Quantas pessoas estão com você: "))

nome = input("Informe seu nome: ")
cpf = input("Informe seu CPF:")
quarto.append(nome)
quarto.append(cpf)

for i in range(quantidade):
  nome = input("Informe o nome de seus companheiros de quarto: ")
  cpf = input("Informe o CPF:")
  agregados = (nome, cpf)
  quarto.append(agregados)

print(quarto)
```
