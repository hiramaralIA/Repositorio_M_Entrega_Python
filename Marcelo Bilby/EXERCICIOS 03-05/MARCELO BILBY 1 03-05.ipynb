**ESTRUTURA DE REPETIÇÃO**
<h1>FOR</h1>

```python
produtos = ['coca', 'pepsi', 'guaraná', 'sprite', 'fanta']
producao = [15000, 12000, 13000, 5000, 250]
```

```python
for i in range(5):
  print(f'Produção do {produtos[i]} é de {producao[i]}')
```

```python
tamanho = len(produtos) 
for i in range(tamanho): 
  print(f'Produção do {produtos[i]} é de {producao[i]}')
```

<h1>WHILE</h1>

```python
vendas = []
```

```python
venda = input("Registre um produto. Para cancelar o registro de um novo produto, basta apertar ENTER com a caixa vazia: ")
while venda != '':
  vendas.append(venda)
  venda = input("Registre um produto. Para cancelar o registro de um novo produto, basta apertar ENTER com a caixa vazia: ")

print(f"Registro finalizado, as vendas cadastradas foram {vendas}")
```
