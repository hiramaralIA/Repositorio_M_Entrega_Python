**Faça uma Função:**

Escrever um algoritmo para uma empresa que decide dar um
reajuste a seus 584 funcionários de acordo com os
seguintes critérios:
a) 50% para aqueles que ganham menos do que três salários mínimos;
b) 20% para aqueles que ganham entre três até dez salários mínimos;
c) 15% para aqueles que ganham acima de dez até vinte salários mínimos;
d) 10% para os demais funcionários.
Leia o nome do funcionário, seu salário e o valor do salário mínimo.
Calcule o seu novo salário reajustado. Escrever o
nome do funcionário, o reajuste e seu novo salário. Calcule quanto à
empresa vai aumentar sua folha de pagamento.


```python
class Funcionario:
  
  funcionario = []
  sumSalario = 0.00
  sumSalarioAtual = 0.00

  def __init__(self, salarioMinimo, nome, salario):
    self.nome = nome
    self.salario = salario
    self.salarioMinimo = salarioMinimo
    self.novoSalario = 0.00
    self.perc = 0

  
  def cadFuncionario(self):
    informacao = (self.nome, self.perc, self.salario, self.novoSalario)
    self.funcionario.append(informacao)

  def reajuste(self):
    if (self.salario < 3 * self.salarioMinimo): 
      self.perc = 50
      self.novoSalario = self.salario + (self.salario * 0.5)
    
    elif (self.salario >= 3 * self.salarioMinimo and self.salario <= 10 * self.salarioMinimo):
          self.perc = 20
          self.novoSalario = self.salario + (self.salario * 0.2)

    elif (self.salario > 10 * self.salarioMinimo and self.salario <= 20 * self.salarioMinimo):
          self.perc = 15
          self.novoSalario = self.salario + (self.salario * 0.15)
    else: 
          self.perc = 10
          self.novoSalario = self.salario + (self.salario * 0.1)


  def getFuncionario(self):
    print("NOME             REAJUSTE(%)  SALARIO ANTERIOR    NOVO SALARIO")
    print("==============================================================")
    for i in range(0, len(self.funcionario)):
      print(f"{self.funcionario[i][0]:20}   {self.funcionario[i][1]:4}%  {self.funcionario[i][2]:17} {self.funcionario[i][3]:14}")
      self.sumSalario += self.funcionario[i][2]
      self.sumSalarioAtual += self.funcionario[i][3]
    print("==============================================================")
    print(f"FOLHA ANTERIOR: {self.sumSalario}     FOLHA ATUAL: {self.sumSalarioAtual}")
     
numeroDeFuncionario = int(input("Informe o número de funcionários a cadastrar: "))
salarioMinimo = float(input("Informe o valor do Salario Minimo: "))

for i in range(0, numeroDeFuncionario):
  
  nome = input("Informe o nome do funcionário: ")
  salario = float(input("Informe o salário do funcionário: "))
  cadastroFuncionario = Funcionario(salarioMinimo, nome, salario)
  cadastroFuncionario.reajuste()
  cadastroFuncionario.cadFuncionario()

print("\n")
cadastroFuncionario.getFuncionario()
```
