**3 - FORAM ANOTADAS AS IDADES E ALTURAS DE 30 ALUNOS. FAÇA UM PROGRAMA QUE DETERMINE QUANTOS ALUNOS COM MAIS DE 13 ANOS POSSUEM ALTURA INFERIOR A MÉDIA DE ALTURA DESSES ALUNOS** 

```python
idades = [32,89,65,34,22,34,76,12,13,29,28,54,24]
alturas = [1.87,1.65,1.23,1.89,1.90,1.32,1.56,1.87,1.04,1.21,1.06,1.76,1.58]
```

```python
i = 0
soma = 0

while i < len(alturas):
    soma += alturas[i]
    i += 1
media = soma / len(alturas)

c = 0
quantidade = 0
while c < len(idades):
    if(idades[c] > 13 and alturas[c] < media):
        quantidade += 1
    c += 1
print(quantidade)
```

**4 - Faça um programa que receba a temperatura média de cada mês do ano e armazene em uma lista. Após isto, calcule a média anual das temperaturas e mostre todas as temperaturas acima da média anual, e em que mês elas ocorreram(mostrar o mês por extenso: 1 - Janeiro, 2 - Fevereiro,...)** 

```python
temperaturaM = {}
meses = ['janeiro',
         'fevereiro',
         'março',
         'abril',
         'maio',
         'junho', 
         'julho',
         'agosto',
         'setembro',
         'outubro',
         'novembro',
         'dezembro']
```

```python
soma = 0
media = 0
for i in range(0, 12):
  mes = float(input(f"Informe a média da temperatura em {meses[i]}: "))
  temperaturaM[meses[i]] = mes
  soma += mes

media = soma / 12

for chave in temperaturaM:
  if temperaturaM[chave] > media:
    print(f"{chave} = {temperaturaM[chave]} C°")
```

**5 - Utilizando uma lista faça um programa que faça 5 perguntas para uma pessoa sobre um crime. As perguntas são:
a) "Telefonou para a vitima?"
b) "Esteve no local do crime?"
c) "Mora perto da vítima?"
d) "Devia para a vítima?"
e "Já trabalhou com a vítima?"**

O programa deve no final emitir uma classificação sobre a participação da pessoa no crime. Se a pessoa responder positivamente a 2 questões ela deve ser classificada como "Suspeita", entre 3 e 4 como "Cúmplice" e 5 como "Assassino". Caso contrário, ele será classificado como "Inocente".

```python
perguntas = ['Telefonou para a vítima?', 'Esteve no local do crime?', 'Mora perto da vítima',
 'Devia para a vítima?', 'Ja trabalhou com a vítima?']
```

```python
contador = 0
for i in range(0, 5):
  resp = input(f"{perguntas[i]} Sim/Não: ")
  if resp.lower() == 'sim':
    contador += 1
if contador == 2:
  print("Suspeita")
else:
  if contador >= 3 and contador <= 4:
    print("Cúmplice")
  else:
    if contador == 5:
      print("Assassino")
```

**8 - Crie um programa que cadastre informações de várias pessoa(nome, idade, cpf, e depois coloque em um dicionário. Depois remova todas as pessoas menores de 18 anos do dicionário e coloque em outro dicionário.**

```python
cadastro = {}
outroDicionário = {}
menor = []
```

```python
s_ou_n='s'
while s_ou_n=='s':
    nome= input('digite o nome:')
    idade= input('digite a idade: ')
    cpf= input('digite o CPF: ')
    cadastro[nome]=[nome,idade,cpf]
    s_ou_n= input('deseja add outro nome? (s/n)')

for chave in cadastro:
  if int(cadastro[chave][1]) < 18:
    menor.append(chave)

for i in menor:
  del cadastro[i]

outroDicionário = cadastro.copy() 
print(outroDicionário)
```
